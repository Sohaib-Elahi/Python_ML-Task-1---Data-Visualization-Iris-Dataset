# Task 1: Exploring and Visualizing a Simple Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("Dataset\Iris.csv")
```

```python
iris.head()
```

```python
print("Iris Data Target Column\n")
print(iris.head())
print("********************\n")
print("Shape Of Iris Data\n")
print(iris.shape)
print("********************\n")
print("Info")
iris.info()
print("********************\n")
print("Iris Data Description\n")
iris.describe()
```

```python
iris.plot(kind='scatter', x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python
sns.FacetGrid(iris, hue="Species").map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
plt.show()
```

```python
sns.FacetGrid(iris, hue="Species", height=2).map(sns.histplot, "PetalLengthCm").add_legend()
sns.FacetGrid(iris, hue="Species", height=2).map(sns.histplot, "PetalWidthCm").add_legend()
sns.FacetGrid(iris, hue="Species", height=2).map(sns.histplot, "SepalLengthCm").add_legend()
sns.FacetGrid(iris, hue="Species", height=2).map(sns.histplot, "SepalWidthCm").add_legend()
plt.show()
```

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")

plt.show()
```

```python
ax = sns.boxplot(x="Species", y="PetalWidthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalWidthCm", data=iris, jitter=True, edgecolor="gray")

plt.show()
```

```python
ax = sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="SepalLengthCm", data=iris, jitter=True, edgecolor="gray")

plt.show()
```

```python
ax = sns.boxplot(x="Species", y="SepalWidthCm", data=iris)
ax = sns.stripplot(x="Species", y="SepalWidthCm", data=iris, jitter=True, edgecolor="gray")

plt.show()
```
